### Проверим, что эффективность обогащения рассчитана правильно

#### Вычислим эффективность обогащения на обучающей выборке для признака `rougher.output.recovery`

```python
# Рассчитаем эффективность обогащения самостоятельно по формуле указанной в описании проекта
data_train_enrichment = pd.read_csv('/datasets/gold_recovery_train_new.csv')
C = data_train_enrichment['rougher.output.concentrate_au'] # доля золота в концентрате после флотации/очистки
F = data_train_enrichment['rougher.input.feed_au'] # доля золота в сырье/концентрате до флотации/очистки
T = data_train_enrichment['rougher.output.tail_au'] # доля золота в отвальных хвостах после флотации/очистки

data_train_enrichment['enrichment'] = C*(F-T)/(F*(C-T))*100
```

### Напишем функции для вычисления итоговой sMAPE используя формулы метрик качества из описания проекта

#### Напишем функцию для вычисления итоговой sMAPE по первой формуле метрик качества из описания проекта

<img src="https://pictures.s3.yandex.net/resources/smape_1576239058.jpg"/>

*Функция принимает на вход значение целевого признака и значение предсказания, на выходе выдает метрику расчитанную по первой формуле.*

```python
def smape(target, predict):
    smape = (1/len(target))*np.sum((np.abs(target - predict))/( (np.abs(target)+np.abs(predict))/2))*100
    return smape
```

```python
smape_scorer = make_scorer(smape, greater_is_better = False)
```

#### Напишем функцию для вычисления итоговой sMAPE по второй формуле метрик качества из описания проекта

<img src= "https://pictures.s3.yandex.net/resources/_smape_1576239054.jpg"/>

*Функция принимает на вход значение целевых признаков и значение предсказаний, на выходе выдает метрику расчитанную по второй формуле.*

```python
def final_smape(rougher, final):
    final = 0.25*rougher+0.75*final
    return final
```

#### Проведем обучение и оценим качество моделей Линейной регрессии, Решающего дерева и Случайного Леса к задачи регрессии. 

*Выведем метрику проверяемых моделей на кросс-валидации. И в итоге посчитаем взвешенную метрику. Используя инструмент smape_scorer оценим качество, так как это регрессия, параметр должен быть низким (метрика отрицательная, умножим на -1)*

##### Проведем обучение и оценим качество модели Линейной регрессии 

```python
# Испытаем модель LinearRegression
model_lr_rougher = LinearRegression()
smape_rougher = cross_val_score(model_lr_rougher,
                          features_train,
                          target_for_rougher,
                          scoring=smape_scorer,
                          cv=5)
final_smape_rougher = smape_rougher.mean()
model_lr_final = LinearRegression()
smape_final = cross_val_score(model_lr_final,
                        features_train,
                        target_for_final,
                        scoring=smape_scorer,
                        cv=5)
final_smape_final = smape_final.mean()
result_smape = final_smape(final_smape_rougher, final_smape_final)*(-1)
```

##### Проведем обучение и оценим качество модели Решающего дерева к задачи регрессии 

```python
# Испытаем модель DecisionTreeRegressor
best_result = 70
best_param = 0
for n in range(2, 10, 1):
    model_rougher = DecisionTreeRegressor(max_depth=n, random_state=12345)
    smape_rougher = cross_val_score(model_rougher,
                              features_train,
                              target_for_rougher,
                              scoring=smape_scorer,
                              cv=5)
    final_smape_rougher = smape_rougher.mean()

    model_final = DecisionTreeRegressor(max_depth=n, random_state=12345)
    smape_final = cross_val_score(model_final, 
                                  features_train,
                                  target_for_final,
                                  scoring=smape_scorer,
                                  cv=5)
    final_smape_final = smape_final.mean()
    result_smape = final_smape(final_smape_rougher, final_smape_final)*(-1)
    if result_smape < best_result:
        best_result = result_smape
        best_param = n
```
