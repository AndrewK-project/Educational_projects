# Проектная работа по Построению дашбордов в Tableau

# "Исследование с помощью Tableau истории TED-конфренций"

## Описание проекта
В этом проекте нами с помощью Tableau исследована история TED-конфренций. TED (от англ technology, education, design — технологии, образование, дизайн) — некоммерческий фонд, проводящий популярные конференции. Специалисты, выступающий на TED работают в самых разных областях и проводят короткие выступления на актуальные социальные, культурные и научные темы. 

## Описание входных данных
tableau_project_data_1.csv, tableau_project_data_2.csv, tableau_project_data_3.csv — данные выступлений. Все три файла имеют одинаковую структуру полей:
- talk_id — идентификатор выступления
- url — ссылка на запись выступления
- title — название выступления
- description — краткое описание
- film_date — дата записи выступления
- duration — длительность в секундах
- views — количество просмотров
- main_tag — основная категория, к которой относится выступление
- speaker_id — уникальный идентификатор автора выступления
- laughter_count — количество раз, когда аудитория смеялась в ходе выступления
- applause_count — количество раз, когда аудитория апложировала в ходе выступления
- language — язык, на котором велось выступление
- event_id — уникальный идентификатор конференции

tableau_project_event_dict.csv — справочник конференций. Поля таблицы:
- conf_id — уникальный идентификатор конференци
- event — название конференции
- country — страна проведения конференции

tableau_project_speakers_dict.csv — справочник авторов выступления. Поля таблицы:
- author_id — уникальный идентификатор автора выступления
- speaker_name — имя автора
- speaker_occupation — профессиональная область автора
- speaker_description — описание профессиональной дятельности автора.

## Проект был написан согласно инструкции учебной платформы:

### Шаг 1. Загрузите и изучите данные
* Загрузите данные в Tableau.
* С помощью union объедините файлы tableau_project_data в единую таблицу. Объедините таблицы с помощью метода Relationship.
* Изучите состав данных, проверьте типы данных.
* С помощью визуализации изучите, как количество конференций распределено по времени. Определите год, после которого количество выступлений скачкообразно выросло.
* На уровне источника данных настройте фильтр, который исключит все выступления до найденного порогового года.

### Шаг 2. Создайте дашборд «История конференций»
* Создайте лист «Количество конференций по годам».
* Создайте сет с набором данных, в котором будет содержаться топ-10 стран по количеству конференций.
* Создайте сет с набором данных, в котором будет содержаться топ-3 тем по количеству выступлений.
* Создайте линейную диаграмму, на которой по оси Х будут располагаться годы, а по оси Y — число конференций. Оформите график.
* Создайте лист «Количество конференций по странам».
* Создайте горизонтальную столбчатую диаграмму, на которой по оси Х будут располагаться топ-10 стран по количеству конференций, а по оси Y — количество конференций. Оформите график. Вместе со столбцами должно отображаться количество конференций.
* Создайте лист «Конференции по годам: топ-3 тем» с визуализацией типа area charts, у которой по оси X отложены годы проведения конференций, а по оси Y — количество конференций. На графике отобразите топ-10 категорий по количеству выступлений.
* Создайте дашборд «История конференций».
* Задайте фильтр по полю Film Date такой, что он применяется ко всем визуализациям, использующим текущий набор данных.
* В области выводов и наблюдений напишите ответы на вопросы:
В каких странах чаще всего проходили конференции?;
Меняется ли тенденция в изменении количества конференций в зависимости от тематики?;
Как менялось число конференций в разные годы?

### Шаг 3. Создайте дашборд «Тематики выступлений»
* Создайте лист «Выступления по тематикам».
* Создайте столбчатую диаграмму, в которой будут отображаться топ-10 самых популярных тематик выступлений в зависимости от числа просмотров. Не забудьте отсортировать и оформить график. 
* Создайте три вычисляемых поля:
Duration, min — длительность выступления в минутах;
Applause by Duration — отношение числа аплодисментов к длительности выступления в секундах;
Laughter by Duration — отношение числа ситуаций, когда аудитория смеялась к длительности выступления в секундах.
* Создайте лист «Аплодисменты по тематикам» с визуализацией типа box-and-whisker plots. На графике по горизонтальной оси должны быть отложены тематики, по вертикальной переменная Applause by Duration. Каждая точка на графике должна показывать отдельное наблюдение. Сделайте так, чтобы на графике категории были отсортированы по убыванию по медианной Applause by Duration. Создайте сет ТОП-10 тематик по длительности выступлений. Отсортируйте графики по убыванию.
* Создайте копию предыдущего листа. Назовите новый лист «Смех по тематикам». На этой визуализации замените Applause by Duration на Laughter by Duration. Не забудьте оформление и сортировку.
* Создайте новый лист «Связь длительности с просмотрами» с визуализацией типа scatter plots. На графике по оси X должна быть отложена длительность выступления в минутах, по оси Y число просмотров. Каждая точка на графике должна соответствовать одному выступлению. Добавьте на график линию тренда, показывающую зависимость числа просмотров от длительности выступлений.
* Создайте дашборд «Тематики выступлений».
* В области выводов и наблюдений напишите ответы на вопросы:
Какие категории выступлений были наиболее популярны?;
Какие категории чаще всего и реже всего вызывают аплодисменты аудитории? Какое из выступлений вызвало наибольшее число аплодисментов?;
Какие категории чаще всего и реже всего вызывают смех аудитории?; Какое из выступлений оказалось самым смешным?;
Есть ли зависимость между длительностью выступления и количеством просмотров?
